```python
import tensorflow as tf
import numpy as np, pandas as pd, os, cv2, matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

CSV_PATH = "/content/drive/MyDrive/Myositis/S1_FileList_Fold_1.csv"
IMG_DIR  = "/content/drive/MyDrive/Myositis/images/images"

df = pd.read_csv(CSV_PATH)
print("Columns:", df.columns.tolist())
print("Counts:", df['Type'].value_counts().to_dict())


df['Class'] = df['Class'].astype(int)
train_df = df[df['Type']=='Train']
val_df   = df[df['Type']=='Validation']
test_df  = df[df['Type']=='Test']

IMG_SIZE = 224

@tf.function
def _load_and_preprocess(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    img = tf.image.convert_image_dtype(img, tf.float32)  # -> [0,1] float32
    return img, tf.cast(label, tf.int32)

def df_to_dataset(sub_df, shuffle=True, batch_size=16):
    paths  = tf.constant([os.path.join(IMG_DIR, fn) for fn in sub_df['FileNames'].tolist()])
    labels = tf.constant(sub_df['Class'].astype(int).to_numpy())
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(_load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(2048, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = df_to_dataset(train_df, shuffle=True,  batch_size=16)
val_ds   = df_to_dataset(val_df,   shuffle=False, batch_size=16)
test_ds  = df_to_dataset(test_df,  shuffle=False, batch_size=16)

classes = np.unique(train_df['Class'])
cw = compute_class_weight('balanced', classes=classes, y=train_df['Class'])
class_weight = {int(c): float(w) for c, w in zip(classes, cw)}
print("Class weights:", class_weight)

#CNN Model
cnn = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
cnn.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
            loss='binary_crossentropy', metrics=['accuracy'])
cnn.summary()

print("\n🔹 Training Model A: Simple CNN")
hist_cnn = cnn.fit(
    train_ds, validation_data=val_ds,
    epochs=15, class_weight=class_weight,
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-6)]
)
cnn.save("/content/drive/MyDrive/Myositis/model_simpleCNN.h5")

loss, acc = cnn.evaluate(test_ds, verbose=0)
print(f"\n Simple CNN Accuracy: {acc*100:.2f}%")

#VGG16
from tensorflow.keras.applications import VGG16
base = VGG16(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
base.trainable = False  # freeze base

inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
vgg_model = tf.keras.Model(inputs, outputs)

vgg_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
                  loss='binary_crossentropy', metrics=['accuracy'])
vgg_model.summary()

print("\n🔹 Training Model B: VGG16 Transfer Learning")
hist_vgg = vgg_model.fit(
    train_ds, validation_data=val_ds,
    epochs=10, class_weight=class_weight,
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, min_lr=1e-6)]
)
vgg_model.save("/content/drive/MyDrive/Myositis/model_VGG16.h5")

loss, acc = vgg_model.evaluate(test_ds, verbose=0)
print(f"\n VGG16 Transfer Model Accuracy: {acc*100:.2f}%")

#Metrics
def evaluate_and_plot(model, ds, name):
    y_true, y_pred = [], []
    for imgs, labels in ds:
        preds = (model.predict(imgs, verbose=0) > 0.5).astype(int)
        y_true.extend(labels.numpy())
        y_pred.extend(preds.squeeze())
    cm = confusion_matrix(y_true, y_pred)
    print(f"\n{name} Classification Report:\n", classification_report(y_true, y_pred, digits=3))
    plt.imshow(cm, cmap='Blues')
    plt.title(f"{name} Confusion Matrix")
    plt.xlabel("Predicted"); plt.ylabel("True")
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha='center', va='center', color='black')
    plt.show()

evaluate_and_plot(cnn, test_ds, "Simple CNN")
evaluate_and_plot(vgg_model, test_ds, "VGG16 Transfer")
```

```python
# Multiclass VGG16
import tensorflow as tf
import numpy as np, pandas as pd, os, matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

CSV_PATH = "/content/drive/MyDrive/Myositis/S4_FileList_Fold_1.csv"
IMG_DIR  = "/content/drive/MyDrive/Myositis/images/images"

df = pd.read_csv(CSV_PATH)
print("Columns:", df.columns.tolist())
print("Type counts:", df['Type'].value_counts().to_dict())

df['Class'] = df['Class'].astype(int)
train_df = df[df['Type']=='Train'].reset_index(drop=True)
val_df   = df[df['Type']=='Validation'].reset_index(drop=True)
test_df  = df[df['Type']=='Test'].reset_index(drop=True)

num_classes = len(np.unique(train_df['Class']))
assert num_classes >= 3, f"Expected multiclass; found {num_classes}"
print("Detected classes:", sorted(np.unique(train_df['Class']).tolist()))

IMG_SIZE = 224
BATCH    = 16

def _load_and_preprocess(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    return img, tf.cast(label, tf.int32)

def df_to_dataset(sub_df, shuffle=True, batch_size=BATCH):
    paths  = tf.constant([os.path.join(IMG_DIR, fn) for fn in sub_df['FileNames'].tolist()])
    labels = tf.constant(sub_df['Class'].to_numpy().astype(np.int32))
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(_load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(2048, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = df_to_dataset(train_df, shuffle=True)
val_ds   = df_to_dataset(val_df,   shuffle=False)
test_ds  = df_to_dataset(test_df,  shuffle=False)

classes = np.sort(train_df['Class'].unique())
cw_vals = compute_class_weight(class_weight='balanced', classes=classes, y=train_df['Class'])
class_weight = {int(c): float(w) for c, w in zip(classes, cw_vals)}
print("Class weights:", class_weight)


from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input

def df_to_dataset_imagenet(sub_df, shuffle, batch_size=BATCH):
    paths  = tf.constant([os.path.join(IMG_DIR, fn) for fn in sub_df['FileNames'].tolist()])
    labels = tf.constant(sub_df['Class'].to_numpy().astype(np.int32))
    def _load_preproc(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_image(img, channels=3, expand_animations=False)
        img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
        img = tf.cast(img, tf.float32)
        img = tf.keras.applications.vgg16.preprocess_input(img)
        return img, tf.cast(label, tf.int32)
    ds = tf.data.Dataset.from_tensor_slices((paths, labels)).map(_load_preproc, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(2048, reshuffle_each_iteration=True)
    return ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_vgg = df_to_dataset_imagenet(train_df, shuffle=True)
val_vgg   = df_to_dataset_imagenet(val_df,   shuffle=False)
test_vgg  = df_to_dataset_imagenet(test_df,  shuffle=False)

base = VGG16(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
for layer in base.layers:
    layer.trainable = True

inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dense(512,  activation='relu')(x)
x = tf.keras.layers.Dense(256,  activation='relu')(x)
outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)
vgg_mc = tf.keras.Model(inputs, outputs)

vgg_mc.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
vgg_mc.summary()

print("\n🔹 Training: VGG16 (multiclass)")
hist_vgg = vgg_mc.fit(
    train_vgg, validation_data=val_vgg, epochs=12, class_weight=class_weight,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau('val_loss', factor=0.3, patience=2, min_lr=1e-6),
        tf.keras.callbacks.EarlyStopping('val_loss', patience=4, restore_best_weights=True)
    ]
)

loss, acc = vgg_mc.evaluate(test_vgg, verbose=0)
print(f"\n✅ VGG16 (multiclass) — Test Accuracy: {acc*100:.2f}%")

#Metrics
import numpy as np

def evaluate_and_report(model, ds, name):
    y_true, y_pred = [], []
    for imgs, labels in ds:
        probs = model.predict(imgs, verbose=0)
        preds = np.argmax(probs, axis=1)
        y_true.extend(labels.numpy().tolist())
        y_pred.extend(preds.tolist())
    print(f"\n{name} — Classification Report")
    print(classification_report(y_true, y_pred, digits=3))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4.5,4))
    plt.imshow(cm, cmap='Blues')
    plt.title(f"{name} Confusion Matrix")
    plt.xlabel("Predicted"); plt.ylabel("True")
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha='center', va='center', color='black')
    plt.tight_layout(); plt.show()

evaluate_and_report(vgg_mc, test_vgg, "VGG16 (multiclass)")
```

```python
#VGG16 Modified version
import tensorflow as tf
import numpy as np, pandas as pd, os, matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input

CSV_PATH = "/content/drive/MyDrive/Myositis/S1_FileList_Fold_1.csv"   # binary split CSV
IMG_DIR  = "/content/drive/MyDrive/Myositis/images/images"
IMG_SIZE = 224
BATCH    = 16
EPOCHS   = 12

df = pd.read_csv(CSV_PATH)
df['Class'] = df['Class'].astype(int)   # 0/1
train_df = df[df['Type']=='Train'].copy()
val_df   = df[df['Type']=='Validation'].copy()
test_df  = df[df['Type']=='Test'].copy()

print("Split sizes:", {k: len(v) for k,v in {"Train":train_df, "Val":val_df, "Test":test_df}.items()})
print("Train label counts:", train_df['Class'].value_counts().to_dict())

@tf.function
def _load_img(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), antialias=True)
    img = tf.cast(img, tf.float32)
    img = preprocess_input(img)  # ImageNet preprocessing
    return img, tf.cast(label, tf.float32)  # scalar label

def df_to_dataset(sub_df, shuffle=True, batch_size=BATCH):
    paths  = tf.constant([os.path.join(IMG_DIR, fn) for fn in sub_df['FileNames'].tolist()])
    labels = tf.constant(sub_df['Class'].astype(int).to_numpy())
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(_load_img, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(2048, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = df_to_dataset(train_df, shuffle=True)
val_ds   = df_to_dataset(val_df,   shuffle=False)
test_ds  = df_to_dataset(test_df,  shuffle=False)

classes = np.array(sorted(train_df['Class'].unique()))  # [0,1]
cw_vals  = compute_class_weight('balanced', classes=classes, y=train_df['Class'].values)
class_weight = {int(c): float(w) for c, w in zip(classes, cw_vals)}
print("Class weights:", class_weight)

#VGG16 (trainable) + GAP + 1024->512->256 -> 1(sigmoid)

base = VGG16(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
for layer in base.layers:
    layer.trainable = True   # authors kept VGG trainable

inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=True)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dense(512,  activation='relu')(x)
x = tf.keras.layers.Dense(256,  activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
vgg_bin = tf.keras.Model(inputs, outputs)

vgg_bin.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
vgg_bin.summary()

#Training
history = vgg_bin.fit(
    train_ds, validation_data=val_ds, epochs=EPOCHS,
    class_weight=class_weight,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, min_lr=1e-6),
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
    ]
)

vgg_model.save("/content/drive/MyDrive/Myositis/vgg_bestmodel.h5")

#Evaluation
y_true, y_pred = [], []
for imgs, labels in test_ds:
    probs = vgg_bin.predict(imgs, verbose=0).ravel()
    preds = (probs >= 0.5).astype(int)
    y_true.extend(labels.numpy().astype(int))
    y_pred.extend(preds)

print("\nClassification report:\n", classification_report(y_true, y_pred, digits=3))
cm = confusion_matrix(y_true, y_pred)
plt.imshow(cm, cmap='Blues'); plt.title("VGG16 (binary) — Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha='center', va='center', color='black')
plt.show()
```
